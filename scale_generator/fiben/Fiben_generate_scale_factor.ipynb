```python
# =========================================================
# STEP 4 — INSPECT THE REAL FIBEN CSV FILES
# =========================================================
# This step inspects the physical CSV files available in the
# FIBEN data directory and builds a physical manifest for the
# subset selected in SF1.
#
# Outputs:
#   ./fiben_sf_artifacts/manifests/all_csv_files_manifest.csv
#   ./fiben_sf_artifacts/manifests/sf1_physical_manifest.csv
#   ./fiben_sf_artifacts/docs/step4_notes.md
# =========================================================

from pathlib import Path
import pandas as pd
import csv

# ---------------------------------------------------------
# 1) Configure input and output paths
# ---------------------------------------------------------
# IMPORTANT:
# Replace DATA_DIR with the folder that contains the FIBEN
# CSV files extracted from data.zip.
# Example:
#   DATA_DIR = Path("/home/jovyan/privado/fiben/data")
# ---------------------------------------------------------
DATA_DIR = Path("/home/jovyan/privado/fiben/data")

BASE_DIR = Path("./fiben_sf_artifacts")
SPEC_DIR = BASE_DIR / "specification"
MANIFEST_DIR = BASE_DIR / "manifests"
DOCS_DIR = BASE_DIR / "docs"

MANIFEST_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)

print("DATA_DIR =", DATA_DIR.resolve() if DATA_DIR.exists() else DATA_DIR)
print("MANIFEST_DIR =", MANIFEST_DIR.resolve())

# ---------------------------------------------------------
# 2) Load the conceptual-to-physical mapping produced earlier
# ---------------------------------------------------------
mapping_csv_path = SPEC_DIR / "sf1_mapping.csv"

if not mapping_csv_path.exists():
    raise FileNotFoundError(f"Missing mapping file: {mapping_csv_path}")

sf1_mapping = pd.read_csv(mapping_csv_path)
sf1_mapping
```

```python
# ---------------------------------------------------------
# 3) List all CSV files available in the FIBEN data folder
# ---------------------------------------------------------
# This block scans the data folder recursively and creates a
# manifest of every CSV file found.
# ---------------------------------------------------------
if not DATA_DIR.exists():
    raise FileNotFoundError(
        f"DATA_DIR does not exist: {DATA_DIR}\n"
        "Please update DATA_DIR before running this step."
    )

all_csv_files = sorted(DATA_DIR.rglob("*.csv"))

print(f"Total CSV files found: {len(all_csv_files)}")
print("First 10 files:")
for p in all_csv_files[:10]:
    print("-", p.name)
```

```python
# ---------------------------------------------------------
# 4) Helper functions for lightweight CSV inspection
# ---------------------------------------------------------
# These functions:
# - normalize table/file names for matching
# - detect file encoding
# - detect delimiter
# - read the header
# - count lines
#
# The goal is to build a physical manifest without loading the
# full dataset into memory.
# ---------------------------------------------------------

def normalize_name(name: str) -> str:
    """
    Normalize a table or file name to improve matching.
    Example:
      'FINANCIALREPORT.csv' -> 'financialreport'
    """
    name = str(name).strip()
    if name.lower().endswith(".csv"):
        name = name[:-4]
    return name.lower()


def detect_encoding(file_path: Path, encodings=("utf-8", "latin-1", "cp1252")):
    """
    Try a small set of common encodings and return the first one
    that can read the first line successfully.
    """
    for enc in encodings:
        try:
            with open(file_path, "r", encoding=enc, newline="") as f:
                f.readline()
            return enc
        except Exception:
            continue
    return "utf-8"


def detect_delimiter(file_path: Path, encoding: str):
    """
    Detect the CSV delimiter using a small text sample.
    Falls back to comma if detection fails.
    """
    try:
        with open(file_path, "r", encoding=encoding, newline="") as f:
            sample = f.read(4096)
        dialect = csv.Sniffer().sniff(sample, delimiters=[",", ";", "|", "\t"])
        return dialect.delimiter
    except Exception:
        return ","


def read_header(file_path: Path, encoding: str, delimiter: str):
    """
    Read only the header row of the CSV file.
    """
    try:
        with open(file_path, "r", encoding=encoding, newline="") as f:
            reader = csv.reader(f, delimiter=delimiter)
            header = next(reader)
        return header
    except Exception:
        return []


def count_rows(file_path: Path, encoding: str):
    """
    Count the number of data rows in the file.
    Assumes the first line is the header.
    """
    try:
        with open(file_path, "r", encoding=encoding, newline="") as f:
            return max(sum(1 for _ in f) - 1, 0)
    except Exception:
        return None
```

```python
# ---------------------------------------------------------
# 5) Build a physical manifest for all CSV files
# ---------------------------------------------------------
# This manifest is useful for:
# - checking the raw dataset inventory
# - comparing raw files against the SF1 mapping
# - documenting the source package
# ---------------------------------------------------------

all_rows = []

for p in all_csv_files:
    enc = detect_encoding(p)
    delim = detect_delimiter(p, enc)
    header = read_header(p, enc, delim)

    all_rows.append({
        "file_name": p.name,
        "normalized_file_name": normalize_name(p.name),
        "absolute_path": str(p.resolve()),
        "size_bytes": p.stat().st_size,
        "size_mb": round(p.stat().st_size / (1024 * 1024), 4),
        "encoding": enc,
        "delimiter": repr(delim),
        "n_columns": len(header),
        "header_preview": " | ".join(header[:10])
    })

all_csv_manifest = pd.DataFrame(all_rows).sort_values("file_name").reset_index(drop=True)

all_csv_manifest_path = MANIFEST_DIR / "all_csv_files_manifest.csv"
all_csv_manifest.to_csv(all_csv_manifest_path, index=False)

print("Saved:", all_csv_manifest_path.resolve())
all_csv_manifest.head(10)
```

```python
# ---------------------------------------------------------
# 6) Match the SF1 mapping against the real CSV files
# ---------------------------------------------------------
# This block links each selected physical table from the SF1
# mapping to the corresponding CSV file found in DATA_DIR.
# ---------------------------------------------------------

# Normalize mapping table names
sf1_mapping["normalized_physical_table"] = sf1_mapping["physical_table"].apply(normalize_name)

# Prepare a lookup from normalized file name -> file metadata
all_csv_lookup = all_csv_manifest.set_index("normalized_file_name")

matched_rows = []

for _, row in sf1_mapping.iterrows():
    table_name = row["physical_table"]
    norm_table = row["normalized_physical_table"]

    if norm_table in all_csv_lookup.index:
        file_info = all_csv_lookup.loc[norm_table]

        # In case of duplicate normalized names, keep the first row
        if isinstance(file_info, pd.DataFrame):
            file_info = file_info.iloc[0]

        matched_rows.append({
            "concept_name": row["concept_name"],
            "physical_table": table_name,
            "conceptual_block": row["conceptual_block"],
            "role_in_workload": row["role_in_workload"],
            "keep_in_sf1": row["keep_in_sf1"],
            "notes": row["notes"],
            "file_found": True,
            "file_name": file_info["file_name"],
            "absolute_path": file_info["absolute_path"],
            "size_bytes": file_info["size_bytes"],
            "size_mb": file_info["size_mb"],
            "encoding": file_info["encoding"],
            "delimiter": file_info["delimiter"],
            "n_columns": file_info["n_columns"],
            "header_preview": file_info["header_preview"]
        })
    else:
        matched_rows.append({
            "concept_name": row["concept_name"],
            "physical_table": table_name,
            "conceptual_block": row["conceptual_block"],
            "role_in_workload": row["role_in_workload"],
            "keep_in_sf1": row["keep_in_sf1"],
            "notes": row["notes"],
            "file_found": False,
            "file_name": None,
            "absolute_path": None,
            "size_bytes": None,
            "size_mb": None,
            "encoding": None,
            "delimiter": None,
            "n_columns": None,
            "header_preview": None
        })

sf1_physical_manifest = pd.DataFrame(matched_rows)
sf1_physical_manifest
```

```python
# ---------------------------------------------------------
# 7) Add row counts for the matched SF1 files
# ---------------------------------------------------------
# Row counting may take some time, but only for the selected
# SF1 tables. This gives a much more useful physical manifest.
# ---------------------------------------------------------

row_counts = []

for idx, row in sf1_physical_manifest.iterrows():
    if row["file_found"]:
        file_path = Path(row["absolute_path"])
        enc = row["encoding"]
        n_rows = count_rows(file_path, enc)
    else:
        n_rows = None

    row_counts.append(n_rows)

sf1_physical_manifest["n_rows"] = row_counts

sf1_physical_manifest_path = MANIFEST_DIR / "sf1_physical_manifest.csv"
sf1_physical_manifest.to_csv(sf1_physical_manifest_path, index=False)

print("Saved:", sf1_physical_manifest_path.resolve())
sf1_physical_manifest
```

```python
# =========================================================
# STEP 6 — INSPECT THE REAL PHYSICAL SCHEMA OF SF1 TABLES
# =========================================================
# This step reads the selected SF1 CSV files in a lightweight
# way and confirms their real physical schema.
#
# The goal is not to extract SF1 yet.
# The goal is to validate:
# - actual column names
# - actual column counts
# - header consistency
# - whether expected key/filter columns exist
#
# Outputs:
#   ./fiben_sf_artifacts/schema/sf1_table_schemas.csv
#   ./fiben_sf_artifacts/schema/sf1_column_dictionary.csv
#   ./fiben_sf_artifacts/schema/sf1_expected_columns_check.csv
#   ./fiben_sf_artifacts/docs/step6_notes.md
# =========================================================

from pathlib import Path
import pandas as pd
import csv

# ---------------------------------------------------------
# 1) Reuse the folder structure created in previous steps
# ---------------------------------------------------------
BASE_DIR = Path("./fiben_sf_artifacts")
SPEC_DIR = BASE_DIR / "specification"
MANIFEST_DIR = BASE_DIR / "manifests"
DOCS_DIR = BASE_DIR / "docs"
SCHEMA_DIR = BASE_DIR / "schema"

SCHEMA_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)

print("Base directory:", BASE_DIR.resolve())
print("Schema directory:", SCHEMA_DIR.resolve())
print("Documentation directory:", DOCS_DIR.resolve())

# ---------------------------------------------------------
# 2) Load the physical manifest and the materialization plan
# ---------------------------------------------------------
sf1_physical_manifest_path = MANIFEST_DIR / "sf1_physical_manifest.csv"
sf1_materialization_plan_path = SPEC_DIR / "sf1_materialization_plan.csv"

if not sf1_physical_manifest_path.exists():
    raise FileNotFoundError(f"Missing file: {sf1_physical_manifest_path}")

if not sf1_materialization_plan_path.exists():
    raise FileNotFoundError(f"Missing file: {sf1_materialization_plan_path}")

sf1_physical_manifest = pd.read_csv(sf1_physical_manifest_path)
sf1_materialization_plan = pd.read_csv(sf1_materialization_plan_path)

print("\nLoaded previous artifacts successfully.")
print("- sf1_physical_manifest rows:", len(sf1_physical_manifest))
print("- sf1_materialization_plan rows:", len(sf1_materialization_plan))
```

```python
# ---------------------------------------------------------
# 3) Helper functions for lightweight schema inspection
# ---------------------------------------------------------
# These functions avoid loading entire CSV files into memory.
# They only read:
# - the header
# - the first few rows
# ---------------------------------------------------------

def normalize_name(name: str) -> str:
    """
    Normalize names to support case-insensitive and punctuation-
    insensitive matching.
    """
    if pd.isna(name):
        return None
    name = str(name).strip().lower()
    return "".join(ch for ch in name if ch.isalnum() or ch == "_")


def safe_read_header(file_path: Path, encoding: str, delimiter: str):
    """
    Read the header row of a CSV file.
    Returns a list of column names.
    """
    with open(file_path, "r", encoding=encoding, newline="") as f:
        reader = csv.reader(f, delimiter=delimiter)
        header = next(reader)
    return header


def safe_read_sample_rows(file_path: Path, encoding: str, delimiter: str, n_rows: int = 3):
    """
    Read a small number of sample rows from the CSV file.
    This is useful only for schema sanity checks and lightweight previews.
    """
    rows = []
    with open(file_path, "r", encoding=encoding, newline="") as f:
        reader = csv.reader(f, delimiter=delimiter)
        _ = next(reader, None)  # skip header
        for _ in range(n_rows):
            row = next(reader, None)
            if row is None:
                break
            rows.append(row)
    return rows


def short_preview_from_row(row, max_items=5, max_chars=120):
    """
    Build a short textual preview from a CSV row.
    """
    if row is None:
        return None
    preview = " | ".join(str(x) for x in row[:max_items])
    if len(preview) > max_chars:
        preview = preview[:max_chars] + "..."
    return preview
```

```python
# ---------------------------------------------------------
# 4) Build the table-level schema summary for all matched SF1 files
# ---------------------------------------------------------
# For each selected SF1 file, collect:
# - real column names
# - real number of columns
# - a small row preview
# - file-level physical metadata
# ---------------------------------------------------------

schema_rows = []

matched_tables = sf1_physical_manifest[sf1_physical_manifest["file_found"] == True].copy()

for _, row in matched_tables.iterrows():
    file_path = Path(row["absolute_path"])
    encoding = row["encoding"]
    delimiter = row["delimiter"]

    # The delimiter was stored as repr(delim), so we convert it back.
    # Example: "',' -> ","
    delimiter = delimiter.strip("'")

    header = safe_read_header(file_path, encoding, delimiter)
    sample_rows = safe_read_sample_rows(file_path, encoding, delimiter, n_rows=3)

    schema_rows.append({
        "physical_table": row["physical_table"],
        "concept_name": row["concept_name"],
        "file_name": row["file_name"],
        "absolute_path": str(file_path),
        "encoding": encoding,
        "delimiter": delimiter,
        "n_columns_actual": len(header),
        "columns_joined": " | ".join(header),
        "sample_row_1": short_preview_from_row(sample_rows[0]) if len(sample_rows) > 0 else None,
        "sample_row_2": short_preview_from_row(sample_rows[1]) if len(sample_rows) > 1 else None,
        "sample_row_3": short_preview_from_row(sample_rows[2]) if len(sample_rows) > 2 else None,
        "n_rows": row.get("n_rows", None),
        "size_mb": row.get("size_mb", None)
    })

sf1_table_schemas = pd.DataFrame(schema_rows).sort_values("physical_table").reset_index(drop=True)

sf1_table_schemas_path = SCHEMA_DIR / "sf1_table_schemas.csv"
sf1_table_schemas.to_csv(sf1_table_schemas_path, index=False)

print("Saved:", sf1_table_schemas_path.resolve())
sf1_table_schemas
```

```python
# ---------------------------------------------------------
# 5) Build the column dictionary for all matched SF1 tables
# ---------------------------------------------------------
# This creates one row per column, which is useful later for:
# - key matching
# - writing extraction scripts
# - documenting the real schema in the methodology
# ---------------------------------------------------------

column_rows = []

for _, row in matched_tables.iterrows():
    file_path = Path(row["absolute_path"])
    encoding = row["encoding"]
    delimiter = row["delimiter"].strip("'")

    header = safe_read_header(file_path, encoding, delimiter)

    for pos, col in enumerate(header, start=1):
        column_rows.append({
            "physical_table": row["physical_table"],
            "concept_name": row["concept_name"],
            "file_name": row["file_name"],
            "ordinal_position": pos,
            "column_name": col,
            "normalized_column_name": normalize_name(col)
        })

sf1_column_dictionary = pd.DataFrame(column_rows).sort_values(
    ["physical_table", "ordinal_position"]
).reset_index(drop=True)

sf1_column_dictionary_path = SCHEMA_DIR / "sf1_column_dictionary.csv"
sf1_column_dictionary.to_csv(sf1_column_dictionary_path, index=False)

print("Saved:", sf1_column_dictionary_path.resolve())
sf1_column_dictionary.head(20)
```

```python
# ---------------------------------------------------------
# 6) Validate expected primary/filter columns against real headers
# ---------------------------------------------------------
# This block checks whether the columns expected in the
# materialization plan are actually present in the real CSV files.
#
# Important:
# - primary keys are checked directly
# - simple expected filter columns are checked directly
# - derived columns are marked as "derived" and not checked here
# ---------------------------------------------------------

def is_direct_column_reference(value):
    """
    Return True if the value looks like a direct column name.
    Complex expressions such as:
      'derived from ...'
      'K_A ∪ K_B'
      None
    are not checked as direct physical columns.
    """
    if pd.isna(value):
        return False
    value = str(value).strip().lower()
    if value == "":
        return False
    if "derived" in value:
        return False
    if "∪" in value:
        return False
    if " " in value:
        return False
    return True


# Build a lookup: table -> set of normalized actual columns
actual_columns_lookup = {}
for table_name, group in sf1_column_dictionary.groupby("physical_table"):
    actual_columns_lookup[table_name] = set(group["normalized_column_name"].dropna().tolist())

check_rows = []

for _, row in sf1_materialization_plan.iterrows():
    table_name = row["physical_table"]
    actual_cols = actual_columns_lookup.get(table_name, set())

    expected_pk = row["expected_primary_key"]
    expected_filter = row["expected_filter_column"]

    expected_pk_norm = normalize_name(expected_pk)
    expected_filter_norm = normalize_name(expected_filter) if is_direct_column_reference(expected_filter) else None

    pk_exists = expected_pk_norm in actual_cols if expected_pk_norm else None

    if is_direct_column_reference(expected_filter):
        filter_exists = expected_filter_norm in actual_cols
        filter_check_mode = "direct"
    else:
        filter_exists = None
        filter_check_mode = "derived_or_not_applicable"

    check_rows.append({
        "physical_table": table_name,
        "expected_primary_key": expected_pk,
        "expected_primary_key_exists": pk_exists,
        "expected_filter_column": expected_filter,
        "expected_filter_column_exists": filter_exists,
        "filter_check_mode": filter_check_mode
    })

sf1_expected_columns_check = pd.DataFrame(check_rows)

sf1_expected_columns_check_path = SCHEMA_DIR / "sf1_expected_columns_check.csv"
sf1_expected_columns_check.to_csv(sf1_expected_columns_check_path, index=False)

print("Saved:", sf1_expected_columns_check_path.resolve())
sf1_expected_columns_check
```

```python
# =========================================================
# STEP 6R-CHECK — VALIDATE THE POSITIONAL SCHEMA
# =========================================================
# This step validates whether the position-based schema
# recovery works correctly before the real SF1 extraction.
#
# Outputs:
#   ./fiben_sf_artifacts/validation/sf1_positional_schema_validation.csv
#   ./fiben_sf_artifacts/validation/sf1_fk_validation_sample.csv
#   ./fiben_sf_artifacts/docs/step6r_check_notes.md
# =========================================================

from pathlib import Path
import pandas as pd

BASE_DIR = Path("./fiben_sf_artifacts")
SCHEMA_DIR = BASE_DIR / "schema"
MANIFEST_DIR = BASE_DIR / "manifests"
DOCS_DIR = BASE_DIR / "docs"
VALIDATION_DIR = BASE_DIR / "validation"

VALIDATION_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)

sf1_physical_manifest_path = MANIFEST_DIR / "sf1_physical_manifest.csv"

if not sf1_physical_manifest_path.exists():
    raise FileNotFoundError(f"Missing file: {sf1_physical_manifest_path}")

sf1_physical_manifest = pd.read_csv(sf1_physical_manifest_path)

print("Loaded validation inputs successfully.")
print("- sf1_physical_manifest rows:", len(sf1_physical_manifest))

# Optional sanity check: list schema files currently available
print("\nSchema files currently available:")
for p in sorted(SCHEMA_DIR.glob("*")):
    print("-", p.name)
```

```python
# ---------------------------------------------------------
# Helper function: load a headerless CSV using the positional schema
# ---------------------------------------------------------

# Reuse the positional schema dictionary created in Step 6R.
# Make sure sf1_positional_schema is already defined in memory.
if "sf1_positional_schema" not in globals():
    raise NameError(
        "sf1_positional_schema is not defined. "
        "Please run Step 6R before running this validation step."
    )

def load_sf1_table_no_header(file_path, table_name, delimiter=",", encoding="utf-8"):
    """
    Read a headerless SF1 CSV file using the assigned column names
    defined by position for the given table.
    """
    assigned_columns = sf1_positional_schema[table_name]

    df = pd.read_csv(
        file_path,
        header=None,
        names=assigned_columns,
        sep=delimiter,
        encoding=encoding
    )
    return df

print("Headerless loader is ready.")
```

```python
# ---------------------------------------------------------
# 1) Validate structural consistency table by table
# ---------------------------------------------------------
# For each selected table, check:
# - file exists
# - table can be loaded
# - actual number of columns matches expected number
# - primary key candidate has no nulls
# - primary key candidate has high uniqueness
# ---------------------------------------------------------

validation_rows = []

for _, row in sf1_physical_manifest.iterrows():
    if not row["file_found"]:
        validation_rows.append({
            "physical_table": row["physical_table"],
            "file_found": False,
            "load_ok": False,
            "n_columns_expected": None,
            "n_columns_actual": None,
            "column_count_ok": False,
            "pk_column": None,
            "pk_null_count": None,
            "pk_distinct_count": None,
            "pk_row_count": None,
            "pk_uniqueness_ratio": None,
            "pk_plausible": False,
            "notes": "file not found"
        })
        continue

    table_name = row["physical_table"]
    file_path = Path(row["absolute_path"])
    encoding = row["encoding"]
    delimiter = row["delimiter"].strip("'")

    expected_columns = sf1_positional_schema[table_name]
    expected_pk = expected_columns[0]

    try:
        df = load_sf1_table_no_header(
            file_path=file_path,
            table_name=table_name,
            delimiter=delimiter,
            encoding=encoding
        )

        n_columns_expected = len(expected_columns)
        n_columns_actual = len(df.columns)
        column_count_ok = (n_columns_expected == n_columns_actual)

        pk_null_count = int(df[expected_pk].isna().sum())
        pk_distinct_count = int(df[expected_pk].nunique(dropna=True))
        pk_row_count = int(len(df))
        pk_uniqueness_ratio = round(pk_distinct_count / pk_row_count, 6) if pk_row_count > 0 else None

        # A simple plausibility rule:
        # - no nulls in PK
        # - at least 95% uniqueness
        pk_plausible = (pk_null_count == 0) and (pk_uniqueness_ratio is not None) and (pk_uniqueness_ratio >= 0.95)

        validation_rows.append({
            "physical_table": table_name,
            "file_found": True,
            "load_ok": True,
            "n_columns_expected": n_columns_expected,
            "n_columns_actual": n_columns_actual,
            "column_count_ok": column_count_ok,
            "pk_column": expected_pk,
            "pk_null_count": pk_null_count,
            "pk_distinct_count": pk_distinct_count,
            "pk_row_count": pk_row_count,
            "pk_uniqueness_ratio": pk_uniqueness_ratio,
            "pk_plausible": pk_plausible,
            "notes": ""
        })

    except Exception as e:
        validation_rows.append({
            "physical_table": table_name,
            "file_found": True,
            "load_ok": False,
            "n_columns_expected": len(expected_columns),
            "n_columns_actual": None,
            "column_count_ok": False,
            "pk_column": expected_pk,
            "pk_null_count": None,
            "pk_distinct_count": None,
            "pk_row_count": None,
            "pk_uniqueness_ratio": None,
            "pk_plausible": False,
            "notes": f"{type(e).__name__}: {e}"
        })

sf1_positional_schema_validation = pd.DataFrame(validation_rows)

validation_path = VALIDATION_DIR / "sf1_positional_schema_validation.csv"
sf1_positional_schema_validation.to_csv(validation_path, index=False)

print("Saved:", validation_path.resolve())
sf1_positional_schema_validation
```

```python
# ---------------------------------------------------------
# 3) Validate a sample of the main foreign key relationships
# ---------------------------------------------------------
# This is a lightweight relational sanity check.
# It validates whether important foreign key values from one
# table appear in the referenced parent table.
# ---------------------------------------------------------

# Define a small set of key relations to validate
fk_checks = [
    ("CORPORATION", "ISCLASSIFIEDBY", "INDUSTRYSECTORCLASSIFIER", "INDUSTRYSECTORCLASSIFIERID"),
    ("CORPORATION", "ISDOMICILEDIN", "COUNTRY", "COUNTRYID"),
    ("FINANCIALREPORT", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("ELEMENTSOFFINANCIALREPORT", "ISMEMBEROF", "FINANCIALREPORT", "FINANCIALREPORTID"),
    ("SECURITY", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("HOLDING", "ISHELDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("HOLDING", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "ISFACILITATEDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("SECURITIESTRANSACTION", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "HASPRICE", "MONETARYAMOUNT", "MONETARYAMOUNTID"),
]

# Load all selected tables once
loaded_tables = {}

for _, row in sf1_physical_manifest.iterrows():
    if not row["file_found"]:
        continue

    table_name = row["physical_table"]
    file_path = Path(row["absolute_path"])
    encoding = row["encoding"]
    delimiter = row["delimiter"].strip("'")

    loaded_tables[table_name] = load_sf1_table_no_header(
        file_path=file_path,
        table_name=table_name,
        delimiter=delimiter,
        encoding=encoding
    )

fk_validation_rows = []

for child_table, child_fk, parent_table, parent_pk in fk_checks:
    child_df = loaded_tables[child_table]
    parent_df = loaded_tables[parent_table]

    child_values = set(child_df[child_fk].dropna().astype(str).unique())
    parent_values = set(parent_df[parent_pk].dropna().astype(str).unique())

    missing_values = child_values - parent_values
    missing_count = len(missing_values)

    fk_validation_rows.append({
        "child_table": child_table,
        "child_fk": child_fk,
        "parent_table": parent_table,
        "parent_pk": parent_pk,
        "n_distinct_child_values": len(child_values),
        "n_distinct_parent_values": len(parent_values),
        "missing_distinct_values": missing_count,
        "fk_plausible": (missing_count == 0)
    })

sf1_fk_validation_sample = pd.DataFrame(fk_validation_rows)

fk_validation_path = VALIDATION_DIR / "sf1_fk_validation_sample.csv"
sf1_fk_validation_sample.to_csv(fk_validation_path, index=False)

print("Saved:", fk_validation_path.resolve())
sf1_fk_validation_sample
```

```python
# =========================================================
# STEP 7 — MATERIALIZE SF1
# =========================================================
# This step performs the real construction of SF1.
#
# Main actions:
# - load the selected FIBEN CSV files as headerless tables
# - assign positional column names
# - derive the retained key sets
# - filter all selected tables according to the extraction plan
# - save the materialized SF1 tables as CSV files WITH HEADER
# - generate a manifest and a referential validation report
#
# Outputs:
#   ./fiben_sf_artifacts/sf1_materialized/tables/*.csv
#   ./fiben_sf_artifacts/manifests/sf1_materialized_manifest.csv
#   ./fiben_sf_artifacts/validation/sf1_materialized_fk_validation.csv
#   ./fiben_sf_artifacts/docs/step7_notes.md
# =========================================================

from pathlib import Path
import pandas as pd

BASE_DIR = Path("./fiben_sf_artifacts")
MANIFEST_DIR = BASE_DIR / "manifests"
SCHEMA_DIR = BASE_DIR / "schema"
DOCS_DIR = BASE_DIR / "docs"
VALIDATION_DIR = BASE_DIR / "validation"
SF1_DIR = BASE_DIR / "sf1_materialized"
SF1_TABLES_DIR = SF1_DIR / "tables"

MANIFEST_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)
VALIDATION_DIR.mkdir(parents=True, exist_ok=True)
SF1_TABLES_DIR.mkdir(parents=True, exist_ok=True)

sf1_physical_manifest_path = MANIFEST_DIR / "sf1_physical_manifest.csv"
sf1_positional_column_dictionary_path = SCHEMA_DIR / "sf1_positional_column_dictionary.csv"

if not sf1_physical_manifest_path.exists():
    raise FileNotFoundError(f"Missing file: {sf1_physical_manifest_path}")

if not sf1_positional_column_dictionary_path.exists():
    raise FileNotFoundError(f"Missing file: {sf1_positional_column_dictionary_path}")

sf1_physical_manifest = pd.read_csv(sf1_physical_manifest_path)
sf1_positional_column_dictionary = pd.read_csv(sf1_positional_column_dictionary_path)

print("Loaded Step 7 inputs successfully.")
print("- sf1_physical_manifest rows:", len(sf1_physical_manifest))
print("- sf1_positional_column_dictionary rows:", len(sf1_positional_column_dictionary))
```

```python
# ---------------------------------------------------------
# 1) Reconstruct the positional schema dictionary from CSV
# ---------------------------------------------------------
# This makes Step 7 independent from in-memory variables
# created in previous notebook steps.
# ---------------------------------------------------------

sf1_positional_schema = {}

for table_name, group in sf1_positional_column_dictionary.groupby("physical_table"):
    ordered_cols = (
        group.sort_values("ordinal_position")["assigned_column_name"]
        .tolist()
    )
    sf1_positional_schema[table_name] = ordered_cols

print("Reconstructed positional schema for", len(sf1_positional_schema), "tables.")


def load_sf1_table_no_header(file_path, table_name, delimiter=",", encoding="utf-8"):
    """
    Read a headerless CSV file using the positional schema
    assigned to the given SF1 table.
    """
    assigned_columns = sf1_positional_schema[table_name]

    df = pd.read_csv(
        file_path,
        header=None,
        names=assigned_columns,
        sep=delimiter,
        encoding=encoding
    )
    return df


def as_str_set(series):
    """
    Convert a pandas Series into a clean set of string values.
    This is useful to compare IDs across tables while avoiding
    numeric type mismatches.
    """
    return set(series.dropna().astype(str).tolist())
```

```python
# ---------------------------------------------------------
# 2) Load all selected SF1 tables from the physical manifest
# ---------------------------------------------------------
# Only the files marked as found are loaded.
# ---------------------------------------------------------

tables = {}

matched_tables = sf1_physical_manifest[sf1_physical_manifest["file_found"] == True].copy()

for _, row in matched_tables.iterrows():
    table_name = row["physical_table"]
    file_path = Path(row["absolute_path"])
    encoding = row["encoding"]
    delimiter = str(row["delimiter"]).strip("'")

    df = load_sf1_table_no_header(
        file_path=file_path,
        table_name=table_name,
        delimiter=delimiter,
        encoding=encoding
    )

    tables[table_name] = df

print("Loaded tables:", sorted(tables.keys()))
```

```python
# ---------------------------------------------------------
# 3) Derive the retained key sets
# ---------------------------------------------------------
# SF1 is company-centered and key-driven.
# Key sets are derived according to the materialization plan.
# ---------------------------------------------------------

# Root and helper dimensions
K_COUNTRY = as_str_set(tables["COUNTRY"]["COUNTRYID"])
K_INDUSTRY = as_str_set(tables["INDUSTRYSECTORCLASSIFIER"]["INDUSTRYSECTORCLASSIFIERID"])
K_CORP = as_str_set(tables["CORPORATION"]["CORPORATIONID"])

# Reporting layer
K_REPORT = as_str_set(
    tables["FINANCIALREPORT"].loc[
        tables["FINANCIALREPORT"]["ISPROVIDEDBY"].astype(str).isin(K_CORP),
        "FINANCIALREPORTID"
    ]
)

K_REPORT_ELEMENT = as_str_set(
    tables["ELEMENTSOFFINANCIALREPORT"].loc[
        tables["ELEMENTSOFFINANCIALREPORT"]["ISMEMBEROF"].astype(str).isin(K_REPORT),
        "ELEMENTSOFFINANCIALREPORTID"
    ]
)

K_STATEMENT_ELEMENT = as_str_set(
    tables["ELEMENTOFFINANCIALSTATEMENT"].loc[
        tables["ELEMENTOFFINANCIALSTATEMENT"]["ELEMENTOFFINANCIALSTATEMENTID"].astype(str).isin(K_REPORT_ELEMENT),
        "ELEMENTOFFINANCIALSTATEMENTID"
    ]
)

K_DISCLOSURE = as_str_set(
    tables["DISCLOSURE"].loc[
        tables["DISCLOSURE"]["DISCLOSUREID"].astype(str).isin(K_REPORT_ELEMENT),
        "DISCLOSUREID"
    ]
)

# Security layer
K_SECURITY = as_str_set(
    tables["SECURITY"].loc[
        tables["SECURITY"]["ISPROVIDEDBY"].astype(str).isin(K_CORP),
        "SECURITYID"
    ]
)

K_LISTED_SECURITY = as_str_set(
    tables["LISTEDSECURITY"].loc[
        tables["LISTEDSECURITY"]["LISTEDSECURITYID"].astype(str).isin(K_SECURITY),
        "LISTEDSECURITYID"
    ]
)

# Portfolio and transaction layer
K_HOLDING = as_str_set(
    tables["HOLDING"].loc[
        tables["HOLDING"]["REFERSTO"].astype(str).isin(K_LISTED_SECURITY),
        "HOLDINGID"
    ]
)

K_TRANSACTION = as_str_set(
    tables["SECURITIESTRANSACTION"].loc[
        tables["SECURITIESTRANSACTION"]["REFERSTO"].astype(str).isin(K_LISTED_SECURITY),
        "SECURITIESTRANSACTIONID"
    ]
)

K_ACCOUNT = set()
K_ACCOUNT |= as_str_set(
    tables["HOLDING"].loc[
        tables["HOLDING"]["HOLDINGID"].astype(str).isin(K_HOLDING),
        "ISHELDBY"
    ]
)
K_ACCOUNT |= as_str_set(
    tables["SECURITIESTRANSACTION"].loc[
        tables["SECURITIESTRANSACTION"]["SECURITIESTRANSACTIONID"].astype(str).isin(K_TRANSACTION),
        "ISFACILITATEDBY"
    ]
)

K_PERSON = set()
K_PERSON |= as_str_set(
    tables["FINANCIALSERVICEACCOUNT"].loc[
        tables["FINANCIALSERVICEACCOUNT"]["FINANCIALSERVICEACCOUNTID"].astype(str).isin(K_ACCOUNT),
        "ISOWNEDBY"
    ]
)
K_PERSON |= as_str_set(
    tables["FINANCIALSERVICEACCOUNT"].loc[
        tables["FINANCIALSERVICEACCOUNT"]["FINANCIALSERVICEACCOUNTID"].astype(str).isin(K_ACCOUNT),
        "ISMANAGEDBY"
    ]
)

K_AMOUNT = set()
K_AMOUNT |= as_str_set(
    tables["LISTEDSECURITY"].loc[
        tables["LISTEDSECURITY"]["LISTEDSECURITYID"].astype(str).isin(K_LISTED_SECURITY),
        "HASLASTTRADEDVALUE"
    ]
)
K_AMOUNT |= as_str_set(
    tables["SECURITIESTRANSACTION"].loc[
        tables["SECURITIESTRANSACTION"]["SECURITIESTRANSACTIONID"].astype(str).isin(K_TRANSACTION),
        "HASPRICE"
    ]
)

# Quick summary
key_sets_summary = pd.DataFrame([
    ["K_COUNTRY", len(K_COUNTRY)],
    ["K_INDUSTRY", len(K_INDUSTRY)],
    ["K_CORP", len(K_CORP)],
    ["K_REPORT", len(K_REPORT)],
    ["K_REPORT_ELEMENT", len(K_REPORT_ELEMENT)],
    ["K_STATEMENT_ELEMENT", len(K_STATEMENT_ELEMENT)],
    ["K_DISCLOSURE", len(K_DISCLOSURE)],
    ["K_SECURITY", len(K_SECURITY)],
    ["K_LISTED_SECURITY", len(K_LISTED_SECURITY)],
    ["K_HOLDING", len(K_HOLDING)],
    ["K_TRANSACTION", len(K_TRANSACTION)],
    ["K_ACCOUNT", len(K_ACCOUNT)],
    ["K_PERSON", len(K_PERSON)],
    ["K_AMOUNT", len(K_AMOUNT)],
], columns=["key_set_name", "n_values"])

key_sets_summary
```

```python
# ---------------------------------------------------------
# 4) Materialize the filtered SF1 tables
# ---------------------------------------------------------
# Helper dimensions are kept complete.
# The remaining tables are filtered by the retained key sets.
# ---------------------------------------------------------

sf1_tables = {}

# Helper dimensions and root
sf1_tables["COUNTRY"] = tables["COUNTRY"].copy()
sf1_tables["INDUSTRYSECTORCLASSIFIER"] = tables["INDUSTRYSECTORCLASSIFIER"].copy()
sf1_tables["CORPORATION"] = tables["CORPORATION"].copy()

# Reporting
sf1_tables["FINANCIALREPORT"] = tables["FINANCIALREPORT"].loc[
    tables["FINANCIALREPORT"]["FINANCIALREPORTID"].astype(str).isin(K_REPORT)
].copy()

sf1_tables["ELEMENTSOFFINANCIALREPORT"] = tables["ELEMENTSOFFINANCIALREPORT"].loc[
    tables["ELEMENTSOFFINANCIALREPORT"]["ELEMENTSOFFINANCIALREPORTID"].astype(str).isin(K_REPORT_ELEMENT)
].copy()

sf1_tables["ELEMENTOFFINANCIALSTATEMENT"] = tables["ELEMENTOFFINANCIALSTATEMENT"].loc[
    tables["ELEMENTOFFINANCIALSTATEMENT"]["ELEMENTOFFINANCIALSTATEMENTID"].astype(str).isin(K_STATEMENT_ELEMENT)
].copy()

sf1_tables["DISCLOSURE"] = tables["DISCLOSURE"].loc[
    tables["DISCLOSURE"]["DISCLOSUREID"].astype(str).isin(K_DISCLOSURE)
].copy()

# Security
sf1_tables["SECURITY"] = tables["SECURITY"].loc[
    tables["SECURITY"]["SECURITYID"].astype(str).isin(K_SECURITY)
].copy()

sf1_tables["LISTEDSECURITY"] = tables["LISTEDSECURITY"].loc[
    tables["LISTEDSECURITY"]["LISTEDSECURITYID"].astype(str).isin(K_LISTED_SECURITY)
].copy()

# Portfolio and transactions
sf1_tables["HOLDING"] = tables["HOLDING"].loc[
    tables["HOLDING"]["HOLDINGID"].astype(str).isin(K_HOLDING)
].copy()

sf1_tables["SECURITIESTRANSACTION"] = tables["SECURITIESTRANSACTION"].loc[
    tables["SECURITIESTRANSACTION"]["SECURITIESTRANSACTIONID"].astype(str).isin(K_TRANSACTION)
].copy()

sf1_tables["FINANCIALSERVICEACCOUNT"] = tables["FINANCIALSERVICEACCOUNT"].loc[
    tables["FINANCIALSERVICEACCOUNT"]["FINANCIALSERVICEACCOUNTID"].astype(str).isin(K_ACCOUNT)
].copy()

sf1_tables["PERSON"] = tables["PERSON"].loc[
    tables["PERSON"]["PERSONID"].astype(str).isin(K_PERSON)
].copy()

sf1_tables["MONETARYAMOUNT"] = tables["MONETARYAMOUNT"].loc[
    tables["MONETARYAMOUNT"]["MONETARYAMOUNTID"].astype(str).isin(K_AMOUNT)
].copy()

# Preview row counts
sf1_row_counts = pd.DataFrame(
    [{"physical_table": t, "n_rows": len(df)} for t, df in sf1_tables.items()]
).sort_values("physical_table").reset_index(drop=True)

sf1_row_counts
```

```python
# ---------------------------------------------------------
# 5) Save the materialized SF1 tables as CSV files WITH HEADER
# ---------------------------------------------------------
# The source package is headerless, but the derived SF1 files
# are saved with explicit headers to improve reproducibility.
# ---------------------------------------------------------

manifest_rows = []

for table_name, df in sf1_tables.items():
    out_path = SF1_TABLES_DIR / f"{table_name}.csv"
    df.to_csv(out_path, index=False, encoding="utf-8")

    manifest_rows.append({
        "physical_table": table_name,
        "output_file": out_path.name,
        "absolute_path": str(out_path.resolve()),
        "n_rows": len(df),
        "n_columns": len(df.columns),
        "size_bytes": out_path.stat().st_size,
        "size_mb": round(out_path.stat().st_size / (1024 * 1024), 4)
    })

sf1_materialized_manifest = pd.DataFrame(manifest_rows).sort_values("physical_table").reset_index(drop=True)

sf1_materialized_manifest_path = MANIFEST_DIR / "sf1_materialized_manifest.csv"
sf1_materialized_manifest.to_csv(sf1_materialized_manifest_path, index=False)

print("Saved:", sf1_materialized_manifest_path.resolve())
sf1_materialized_manifest
```

```python
# ---------------------------------------------------------
# 6) Validate referential consistency of the materialized SF1
# ---------------------------------------------------------

def missing_fk_count(child_df, child_fk, parent_df, parent_pk):
    child_values = set(child_df[child_fk].dropna().astype(str).unique())
    parent_values = set(parent_df[parent_pk].dropna().astype(str).unique())
    return len(child_values - parent_values)

fk_checks_materialized = [
    ("CORPORATION", "ISCLASSIFIEDBY", "INDUSTRYSECTORCLASSIFIER", "INDUSTRYSECTORCLASSIFIERID"),
    ("CORPORATION", "ISDOMICILEDIN", "COUNTRY", "COUNTRYID"),
    ("FINANCIALREPORT", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("ELEMENTSOFFINANCIALREPORT", "ISMEMBEROF", "FINANCIALREPORT", "FINANCIALREPORTID"),
    ("ELEMENTOFFINANCIALSTATEMENT", "ELEMENTOFFINANCIALSTATEMENTID", "ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID"),
    ("DISCLOSURE", "DISCLOSUREID", "ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID"),
    ("SECURITY", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("LISTEDSECURITY", "LISTEDSECURITYID", "SECURITY", "SECURITYID"),
    ("HOLDING", "ISHELDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("HOLDING", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "ISFACILITATEDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("SECURITIESTRANSACTION", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "HASPRICE", "MONETARYAMOUNT", "MONETARYAMOUNTID"),
    ("FINANCIALSERVICEACCOUNT", "ISOWNEDBY", "PERSON", "PERSONID"),
    ("FINANCIALSERVICEACCOUNT", "ISMANAGEDBY", "PERSON", "PERSONID"),
    ("LISTEDSECURITY", "HASLASTTRADEDVALUE", "MONETARYAMOUNT", "MONETARYAMOUNTID"),
]

fk_validation_rows = []

for child_table, child_fk, parent_table, parent_pk in fk_checks_materialized:
    missing_count = missing_fk_count(
        sf1_tables[child_table], child_fk,
        sf1_tables[parent_table], parent_pk
    )

    fk_validation_rows.append({
        "child_table": child_table,
        "child_fk": child_fk,
        "parent_table": parent_table,
        "parent_pk": parent_pk,
        "missing_distinct_values": missing_count,
        "fk_ok": (missing_count == 0)
    })

sf1_materialized_fk_validation = pd.DataFrame(fk_validation_rows)

sf1_materialized_fk_validation_path = VALIDATION_DIR / "sf1_materialized_fk_validation.csv"
sf1_materialized_fk_validation.to_csv(sf1_materialized_fk_validation_path, index=False)

print("Saved:", sf1_materialized_fk_validation_path.resolve())
sf1_materialized_fk_validation
```

```python
# ---------------------------------------------------------
# 3) Helper function to profile relationship cardinalities
# ---------------------------------------------------------
# For a relationship parent -> child, this function computes
# how many child rows each parent row has.
#
# Reported metrics:
# - number of parent keys observed
# - number of distinct child rows
# - min/mean/median/p95/max children per parent
# ---------------------------------------------------------

def relationship_profile(parent_df, parent_pk, child_df, child_fk, relationship_name):
    """
    Profile the cardinality of parent -> child relationship.
    """
    parent_keys = parent_df[parent_pk].dropna().astype(str)
    child_fk_values = child_df[child_fk].dropna().astype(str)

    counts = child_fk_values.value_counts()

    aligned_counts = parent_keys.map(counts).fillna(0)

    profile = {
        "relationship_name": relationship_name,
        "parent_table": parent_df.attrs.get("table_name", ""),
        "parent_pk": parent_pk,
        "child_table": child_df.attrs.get("table_name", ""),
        "child_fk": child_fk,
        "n_parent_rows": int(len(parent_df)),
        "n_distinct_parent_keys": int(parent_keys.nunique()),
        "n_child_rows": int(len(child_df)),
        "n_distinct_child_fk_values": int(child_fk_values.nunique()),
        "min_children_per_parent": float(aligned_counts.min()) if len(aligned_counts) > 0 else None,
        "mean_children_per_parent": round(float(aligned_counts.mean()), 6) if len(aligned_counts) > 0 else None,
        "median_children_per_parent": round(float(aligned_counts.median()), 6) if len(aligned_counts) > 0 else None,
        "p95_children_per_parent": round(float(aligned_counts.quantile(0.95)), 6) if len(aligned_counts) > 0 else None,
        "max_children_per_parent": float(aligned_counts.max()) if len(aligned_counts) > 0 else None,
    }
    return profile
```

```python
# ---------------------------------------------------------
# 1) Define scale factors and fixed dimensions
# ---------------------------------------------------------
# COUNTRY and INDUSTRYSECTORCLASSIFIER are treated as fixed
# dimensions because they act as contextual vocabularies.
# The remaining tables are scalable.
# ---------------------------------------------------------

fixed_tables = {"COUNTRY", "INDUSTRYSECTORCLASSIFIER"}

scale_factors = pd.DataFrame([
    ["SF1", 1, "baseline", True],
    ["SF10", 10, "main_experiment", True],
    ["SF30", 30, "main_experiment", True],
    ["SF100", 100, "optional_stress_test", False]
], columns=[
    "scale_name",
    "scale_multiplier",
    "usage_role",
    "enabled_by_default"
])

scale_factors
```

```python
# ---------------------------------------------------------
# 2) Compute target row counts and estimated sizes by scale
# ---------------------------------------------------------
# Fixed tables keep their SF1 size.
# Scalable tables are multiplied by the scale factor.
# ---------------------------------------------------------

target_rows = []

for _, sf in scale_factors.iterrows():
    scale_name = sf["scale_name"]
    multiplier = sf["scale_multiplier"]

    for _, row in table_summary.iterrows():
        table_name = row["physical_table"]
        source_rows = int(row["n_rows"])
        source_size_mb = float(row["size_mb"])

        if table_name in fixed_tables:
            target_row_count = source_rows
            target_size_mb = source_size_mb
            scaling_mode = "fixed"
        else:
            target_row_count = int(round(source_rows * multiplier))
            target_size_mb = round(source_size_mb * multiplier, 4)
            scaling_mode = "scaled"

        target_rows.append({
            "scale_name": scale_name,
            "scale_multiplier": multiplier,
            "physical_table": table_name,
            "scaling_mode": scaling_mode,
            "sf1_rows": source_rows,
            "target_rows": target_row_count,
            "sf1_size_mb": source_size_mb,
            "estimated_target_size_mb": target_size_mb
        })

sf_targets = pd.DataFrame(target_rows)

sf_targets_path = AUDIT_DIR / "sf_targets.csv"
sf_targets.to_csv(sf_targets_path, index=False)

print("Saved:", sf_targets_path.resolve())
sf_targets.head(20)
```

```python
# =========================================================
# STEP 10A — SYNTHETIC EXPANSION OF THE CORPORATION-ROOTED SIDE
# =========================================================
# This step generates the corporation-rooted synthetic package
# for a target scale factor (e.g., SF10 or SF30).
#
# It includes:
# - fixed dimensions
# - corporations
# - financial reports
# - report elements
# - statement elements
# - disclosures
# - securities
# - listed securities
#
# It does NOT yet generate:
# - persons
# - accounts
# - holdings
# - transactions
# - monetary amounts
#
# Those will be handled in Step 10B.
#
# Outputs:
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/tables/*.csv
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/manifests/corp_rooted_manifest.csv
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/manifests/pending_links.csv
#   ./fiben_sf_artifacts/docs/step10a_notes.md
# =========================================================

from pathlib import Path
import pandas as pd
import shutil

# ---------------------------------------------------------
# 1) Configure the target scale factor
# ---------------------------------------------------------
# Recommended execution order:
# - first run SF10
# - validate the outputs
# - then run SF30
# ---------------------------------------------------------
TARGET_SCALE = "SF30"   # Change to "SF30" later if needed
OVERWRITE_OUTPUT = False
CHUNK_SIZE = 250_000

# ---------------------------------------------------------
# 2) Define input/output folders
# ---------------------------------------------------------
BASE_DIR = Path("./fiben_sf_artifacts")
SF1_TABLES_DIR = BASE_DIR / "sf1_materialized" / "tables"
AUDIT_DIR = BASE_DIR / "audit"
DOCS_DIR = BASE_DIR / "docs"

TARGET_DIR = BASE_DIR / "scaled_corp_rooted" / TARGET_SCALE
TARGET_TABLES_DIR = TARGET_DIR / "tables"
TARGET_MANIFEST_DIR = TARGET_DIR / "manifests"

TARGET_TABLES_DIR.mkdir(parents=True, exist_ok=True)
TARGET_MANIFEST_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)

print("SF1 input directory:", SF1_TABLES_DIR.resolve())
print("Target output directory:", TARGET_DIR.resolve())

# ---------------------------------------------------------
# 3) Load target multiplier from sf_targets.csv
# ---------------------------------------------------------
sf_targets_path = AUDIT_DIR / "sf_targets.csv"

if not sf_targets_path.exists():
    raise FileNotFoundError(f"Missing file: {sf_targets_path}")

sf_targets = pd.read_csv(sf_targets_path)

if TARGET_SCALE not in set(sf_targets["scale_name"]):
    raise ValueError(f"TARGET_SCALE not found in sf_targets.csv: {TARGET_SCALE}")

target_multiplier = int(
    sf_targets.loc[sf_targets["scale_name"] == TARGET_SCALE, "scale_multiplier"].iloc[0]
)

if target_multiplier < 1:
    raise ValueError("Target multiplier must be >= 1")

replica_count = target_multiplier - 1

print(f"Target scale: {TARGET_SCALE}")
print(f"Multiplier: {target_multiplier}")
print(f"Extra synthetic replicas per SF1 row-set: {replica_count}")
```

```python
# ---------------------------------------------------------
# 4) Define fixed and corporation-rooted tables
# ---------------------------------------------------------
fixed_tables = [
    "COUNTRY",
    "INDUSTRYSECTORCLASSIFIER"
]

corp_rooted_tables = [
    "CORPORATION",
    "FINANCIALREPORT",
    "ELEMENTSOFFINANCIALREPORT",
    "ELEMENTOFFINANCIALSTATEMENT",
    "DISCLOSURE",
    "SECURITY",
    "LISTEDSECURITY"
]

all_step10a_tables = fixed_tables + corp_rooted_tables

# ---------------------------------------------------------
# 5) Prevent accidental duplicate appends
# ---------------------------------------------------------
existing_csvs = list(TARGET_TABLES_DIR.glob("*.csv"))

if existing_csvs and not OVERWRITE_OUTPUT:
    raise FileExistsError(
        f"Output directory already contains CSV files: {TARGET_TABLES_DIR}\n"
        "Set OVERWRITE_OUTPUT=True if you want to recreate the package."
    )

if existing_csvs and OVERWRITE_OUTPUT:
    for p in existing_csvs:
        p.unlink()

print("Step 10A tables:")
for t in all_step10a_tables:
    print("-", t)
```

```python
# ---------------------------------------------------------
# 6) Helper functions
# ---------------------------------------------------------

def read_sf1_table(table_name, dtype=str, keep_default_na=False):
    """
    Read one SF1 table (saved with header) from the materialized SF1 package.
    """
    file_path = SF1_TABLES_DIR / f"{table_name}.csv"
    if not file_path.exists():
        raise FileNotFoundError(f"Missing SF1 table: {file_path}")

    return pd.read_csv(file_path, dtype=dtype, keep_default_na=keep_default_na)


def make_synth_id(series, scale_name, table_code, replica_index):
    """
    Build deterministic synthetic IDs from source IDs.
    Example:
      1000 -> SF10_CORP_R01_1000
    """
    return series.astype(str).map(lambda x: f"{scale_name}_{table_code}_R{replica_index:02d}_{x}")


def mutate_text_value(series, scale_name, replica_index):
    """
    Add a readable synthetic suffix to natural textual identifiers
    such as ticker symbols, legal names, or report identifiers.
    Empty values remain empty.
    """
    series = series.astype(str)
    return series.map(
        lambda x: x if x == "" else f"{x}__{scale_name}_R{replica_index:02d}"
    )


def append_df_to_csv(df, out_path):
    """
    Append a dataframe to an existing CSV file without writing the header.
    """
    df.to_csv(out_path, mode="a", index=False, header=False, encoding="utf-8")


def copy_base_csv(table_name):
    """
    Copy the original SF1 CSV as the starting point of the scaled package.
    """
    src = SF1_TABLES_DIR / f"{table_name}.csv"
    dst = TARGET_TABLES_DIR / f"{table_name}.csv"
    shutil.copy2(src, dst)
    return dst
```

```python
# ---------------------------------------------------------
# 7) Copy the base SF1 package for the selected Step 10A tables
# ---------------------------------------------------------
# The scaled package starts with the original SF1 rows and
# then appends synthetic replicas.
# ---------------------------------------------------------

copied_files = []

for table_name in all_step10a_tables:
    dst = copy_base_csv(table_name)
    copied_files.append((table_name, dst.name))

copied_files_df = pd.DataFrame(copied_files, columns=["physical_table", "copied_file"])
copied_files_df
```

```python
# ---------------------------------------------------------
# 8) Load the smaller source tables fully in memory
# ---------------------------------------------------------
# These tables are used repeatedly across replicas and are
# small enough to be handled as full dataframes.
# ---------------------------------------------------------

CORPORATION = read_sf1_table("CORPORATION")
FINANCIALREPORT = read_sf1_table("FINANCIALREPORT")
SECURITY = read_sf1_table("SECURITY")
LISTEDSECURITY = read_sf1_table("LISTEDSECURITY")

print("Loaded corporation-rooted small tables successfully.")
print("- CORPORATION:", len(CORPORATION))
print("- FINANCIALREPORT:", len(FINANCIALREPORT))
print("- SECURITY:", len(SECURITY))
print("- LISTEDSECURITY:", len(LISTEDSECURITY))
```

```python
# ---------------------------------------------------------
# 9) Generate and append synthetic replicas for the small
#    corporation-rooted tables
# ---------------------------------------------------------
# This block handles:
# - CORPORATION
# - FINANCIALREPORT
# - SECURITY
# - LISTEDSECURITY
#
# The large reporting tables will be handled chunk-wise next.
# ---------------------------------------------------------

corporation_out = TARGET_TABLES_DIR / "CORPORATION.csv"
financialreport_out = TARGET_TABLES_DIR / "FINANCIALREPORT.csv"
security_out = TARGET_TABLES_DIR / "SECURITY.csv"
listedsecurity_out = TARGET_TABLES_DIR / "LISTEDSECURITY.csv"

replica_stats = []

for rep in range(1, replica_count + 1):
    print(f"\nGenerating small-table replica {rep}/{replica_count} for {TARGET_SCALE}...")

    # -------------------------
    # CORPORATION
    # -------------------------
    corp_clone = CORPORATION.copy()
    corp_id_map = dict(
        zip(
            CORPORATION["CORPORATIONID"].astype(str),
            make_synth_id(CORPORATION["CORPORATIONID"], TARGET_SCALE, "CORP", rep)
        )
    )

    corp_clone["CORPORATIONID"] = corp_clone["CORPORATIONID"].astype(str).map(corp_id_map)

    if "HASTICKERSYMBOL" in corp_clone.columns:
        corp_clone["HASTICKERSYMBOL"] = mutate_text_value(corp_clone["HASTICKERSYMBOL"], TARGET_SCALE, rep)

    if "HASLEGALNAME" in corp_clone.columns:
        corp_clone["HASLEGALNAME"] = mutate_text_value(corp_clone["HASLEGALNAME"], TARGET_SCALE, rep)

    append_df_to_csv(corp_clone, corporation_out)

    # -------------------------
    # FINANCIALREPORT
    # -------------------------
    report_clone = FINANCIALREPORT.copy()
    report_id_map = dict(
        zip(
            FINANCIALREPORT["FINANCIALREPORTID"].astype(str),
            make_synth_id(FINANCIALREPORT["FINANCIALREPORTID"], TARGET_SCALE, "FREP", rep)
        )
    )

    report_clone["FINANCIALREPORTID"] = report_clone["FINANCIALREPORTID"].astype(str).map(report_id_map)
    report_clone["ISPROVIDEDBY"] = report_clone["ISPROVIDEDBY"].astype(str).map(corp_id_map)

    if "HASUNIQUEIDENTIFIER" in report_clone.columns:
        report_clone["HASUNIQUEIDENTIFIER"] = mutate_text_value(report_clone["HASUNIQUEIDENTIFIER"], TARGET_SCALE, rep)

    append_df_to_csv(report_clone, financialreport_out)

    # -------------------------
    # SECURITY
    # -------------------------
    security_clone = SECURITY.copy()
    security_id_map = dict(
        zip(
            SECURITY["SECURITYID"].astype(str),
            make_synth_id(SECURITY["SECURITYID"], TARGET_SCALE, "SEC", rep)
        )
    )

    security_clone["SECURITYID"] = security_clone["SECURITYID"].astype(str).map(security_id_map)
    security_clone["ISPROVIDEDBY"] = security_clone["ISPROVIDEDBY"].astype(str).map(corp_id_map)

    append_df_to_csv(security_clone, security_out)

    # -------------------------
    # LISTEDSECURITY
    # -------------------------
    # Preserve the 1:1 mapping between SECURITY and LISTEDSECURITY
    listedsecurity_clone = LISTEDSECURITY.copy()
    listedsecurity_clone["LISTEDSECURITYID"] = listedsecurity_clone["LISTEDSECURITYID"].astype(str).map(security_id_map)

    if "HASTICKERSYMBOL" in listedsecurity_clone.columns:
        listedsecurity_clone["HASTICKERSYMBOL"] = mutate_text_value(
            listedsecurity_clone["HASTICKERSYMBOL"], TARGET_SCALE, rep
        )

    if "HASLEGALNAME" in listedsecurity_clone.columns:
        listedsecurity_clone["HASLEGALNAME"] = mutate_text_value(
            listedsecurity_clone["HASLEGALNAME"], TARGET_SCALE, rep
        )

    # Note:
    # HASLASTTRADEDVALUE is left unchanged in this step.
    # It will be closed later together with MONETARYAMOUNT in Step 10B.
    append_df_to_csv(listedsecurity_clone, listedsecurity_out)

    replica_stats.extend([
        [TARGET_SCALE, rep, "CORPORATION", len(corp_clone)],
        [TARGET_SCALE, rep, "FINANCIALREPORT", len(report_clone)],
        [TARGET_SCALE, rep, "SECURITY", len(security_clone)],
        [TARGET_SCALE, rep, "LISTEDSECURITY", len(listedsecurity_clone)],
    ])

replica_stats_df = pd.DataFrame(replica_stats, columns=[
    "scale_name", "replica_index", "physical_table", "n_rows_appended"
])

replica_stats_df.head(20)
```

```python
# ---------------------------------------------------------
# 10) Generate and append synthetic replicas for the large
#     reporting tables using chunked processing
# ---------------------------------------------------------
# This block handles:
# - ELEMENTSOFFINANCIALREPORT
# - ELEMENTOFFINANCIALSTATEMENT
# - DISCLOSURE
#
# Chunking avoids loading the largest tables fully into memory
# during each replica generation.
# ---------------------------------------------------------

report_elements_src = SF1_TABLES_DIR / "ELEMENTSOFFINANCIALREPORT.csv"
statement_elements_src = SF1_TABLES_DIR / "ELEMENTOFFINANCIALSTATEMENT.csv"
disclosure_src = SF1_TABLES_DIR / "DISCLOSURE.csv"

report_elements_out = TARGET_TABLES_DIR / "ELEMENTSOFFINANCIALREPORT.csv"
statement_elements_out = TARGET_TABLES_DIR / "ELEMENTOFFINANCIALSTATEMENT.csv"
disclosure_out = TARGET_TABLES_DIR / "DISCLOSURE.csv"

for rep in range(1, replica_count + 1):
    print(f"\nGenerating large reporting-table replica {rep}/{replica_count} for {TARGET_SCALE}...")

    # Rebuild report ID map for this replica
    report_id_map = dict(
        zip(
            FINANCIALREPORT["FINANCIALREPORTID"].astype(str),
            make_synth_id(FINANCIALREPORT["FINANCIALREPORTID"], TARGET_SCALE, "FREP", rep)
        )
    )

    # -------------------------
    # ELEMENTSOFFINANCIALREPORT
    # -------------------------
    for chunk in pd.read_csv(report_elements_src, dtype=str, keep_default_na=False, chunksize=CHUNK_SIZE):
        chunk["ELEMENTSOFFINANCIALREPORTID"] = make_synth_id(
            chunk["ELEMENTSOFFINANCIALREPORTID"], TARGET_SCALE, "RELM", rep
        )
        chunk["ISMEMBEROF"] = chunk["ISMEMBEROF"].astype(str).map(report_id_map)

        append_df_to_csv(chunk, report_elements_out)

    # -------------------------
    # ELEMENTOFFINANCIALSTATEMENT
    # -------------------------
    for chunk in pd.read_csv(statement_elements_src, dtype=str, keep_default_na=False, chunksize=CHUNK_SIZE):
        chunk["ELEMENTOFFINANCIALSTATEMENTID"] = make_synth_id(
            chunk["ELEMENTOFFINANCIALSTATEMENTID"], TARGET_SCALE, "RELM", rep
        )

        append_df_to_csv(chunk, statement_elements_out)

    # -------------------------
    # DISCLOSURE
    # -------------------------
    for chunk in pd.read_csv(disclosure_src, dtype=str, keep_default_na=False, chunksize=CHUNK_SIZE):
        chunk["DISCLOSUREID"] = make_synth_id(
            chunk["DISCLOSUREID"], TARGET_SCALE, "RELM", rep
        )

        append_df_to_csv(chunk, disclosure_out)

print("Large reporting-table generation completed.")
```

```python
# ---------------------------------------------------------
# 11) Build the corporation-rooted package manifest
# ---------------------------------------------------------
# This manifest documents the partial scaled package produced
# in Step 10A.
# ---------------------------------------------------------

manifest_rows = []

for table_name in all_step10a_tables:
    file_path = TARGET_TABLES_DIR / f"{table_name}.csv"

    # Expected row count:
    # - fixed tables stay unchanged
    # - corporation-rooted tables are multiplied by target_multiplier
    source_rows = len(read_sf1_table(table_name))

    if table_name in fixed_tables:
        target_rows = source_rows
        scaling_mode = "fixed"
    else:
        target_rows = source_rows * target_multiplier
        scaling_mode = "scaled"

    manifest_rows.append({
        "scale_name": TARGET_SCALE,
        "physical_table": table_name,
        "scaling_mode": scaling_mode,
        "sf1_rows": source_rows,
        "expected_rows_after_step10a": target_rows,
        "output_file": file_path.name,
        "absolute_path": str(file_path.resolve()),
        "size_bytes": file_path.stat().st_size,
        "size_mb": round(file_path.stat().st_size / (1024 * 1024), 4)
    })

corp_rooted_manifest = pd.DataFrame(manifest_rows).sort_values("physical_table").reset_index(drop=True)

corp_rooted_manifest_path = TARGET_MANIFEST_DIR / "corp_rooted_manifest.csv"
corp_rooted_manifest.to_csv(corp_rooted_manifest_path, index=False)

print("Saved:", corp_rooted_manifest_path.resolve())
corp_rooted_manifest
```

```python
# =========================================================
# STEP 10B — COMPLETE THE SCALE FACTOR WITH THE ACCOUNT-ROOTED SIDE
# =========================================================
# This step closes the synthetic scale-factor package created
# in Step 10A.
#
# Main actions:
# - rebuild LISTEDSECURITY with synthetic monetary amount links
# - generate synthetic PERSON rows
# - generate synthetic FINANCIALSERVICEACCOUNT rows
# - generate synthetic HOLDING rows
# - generate synthetic SECURITIESTRANSACTION rows
# - generate synthetic MONETARYAMOUNT rows
# - build the final full manifest
# - validate referential consistency of the full scaled package
#
# Outputs:
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/tables/*.csv
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/manifests/final_scaled_manifest.csv
#   ./fiben_sf_artifacts/scaled_corp_rooted/<TARGET_SCALE>/validation/final_scaled_fk_validation.csv
#   ./fiben_sf_artifacts/docs/step10b_notes.md
# =========================================================

from pathlib import Path
import pandas as pd

# ---------------------------------------------------------
# 1) Configure the target scale factor
# ---------------------------------------------------------
TARGET_SCALE = "SF30"   # Change to "SF30" later if needed

BASE_DIR = Path("./fiben_sf_artifacts")
SF1_TABLES_DIR = BASE_DIR / "sf1_materialized" / "tables"
AUDIT_DIR = BASE_DIR / "audit"
DOCS_DIR = BASE_DIR / "docs"

TARGET_DIR = BASE_DIR / "scaled_corp_rooted" / TARGET_SCALE
TARGET_TABLES_DIR = TARGET_DIR / "tables"
TARGET_MANIFEST_DIR = TARGET_DIR / "manifests"
TARGET_VALIDATION_DIR = TARGET_DIR / "validation"

TARGET_MANIFEST_DIR.mkdir(parents=True, exist_ok=True)
TARGET_VALIDATION_DIR.mkdir(parents=True, exist_ok=True)
DOCS_DIR.mkdir(parents=True, exist_ok=True)

sf_targets_path = AUDIT_DIR / "sf_targets.csv"
corp_rooted_manifest_path = TARGET_MANIFEST_DIR / "corp_rooted_manifest.csv"

if not sf_targets_path.exists():
    raise FileNotFoundError(f"Missing file: {sf_targets_path}")

if not corp_rooted_manifest_path.exists():
    raise FileNotFoundError(f"Missing file: {corp_rooted_manifest_path}")

sf_targets = pd.read_csv(sf_targets_path)
corp_rooted_manifest = pd.read_csv(corp_rooted_manifest_path)

if TARGET_SCALE not in set(sf_targets["scale_name"]):
    raise ValueError(f"TARGET_SCALE not found in sf_targets.csv: {TARGET_SCALE}")

target_multiplier = int(
    sf_targets.loc[sf_targets["scale_name"] == TARGET_SCALE, "scale_multiplier"].iloc[0]
)

if target_multiplier < 1:
    raise ValueError("Target multiplier must be >= 1")

replica_count = target_multiplier - 1

print("Target scale:", TARGET_SCALE)
print("Target directory:", TARGET_DIR.resolve())
print("Multiplier:", target_multiplier)
print("Extra synthetic replicas:", replica_count)
```

```python
# ---------------------------------------------------------
# 2) Helper functions
# ---------------------------------------------------------

def read_sf1_table(table_name, dtype=str, keep_default_na=False):
    """
    Read one materialized SF1 table with header.
    """
    file_path = SF1_TABLES_DIR / f"{table_name}.csv"
    if not file_path.exists():
        raise FileNotFoundError(f"Missing SF1 table: {file_path}")

    return pd.read_csv(file_path, dtype=dtype, keep_default_na=keep_default_na)


def read_scaled_table(table_name, dtype=str, keep_default_na=False):
    """
    Read one table from the partial scaled package generated in Step 10A.
    """
    file_path = TARGET_TABLES_DIR / f"{table_name}.csv"
    if not file_path.exists():
        raise FileNotFoundError(f"Missing scaled table: {file_path}")

    return pd.read_csv(file_path, dtype=dtype, keep_default_na=keep_default_na)


def make_synth_id(series, scale_name, table_code, replica_index):
    """
    Build deterministic synthetic IDs.
    """
    return series.astype(str).map(lambda x: f"{scale_name}_{table_code}_R{replica_index:02d}_{x}")


def mutate_text_value(series, scale_name, replica_index):
    """
    Add a readable suffix to text attributes to reduce collisions.
    """
    series = series.astype(str)
    return series.map(lambda x: x if x == "" else f"{x}__{scale_name}_R{replica_index:02d}")


def map_with_fallback(series, mapping):
    """
    Map values through a dictionary while preserving original values
    if they are not found in the mapping.
    """
    s = series.astype(str)
    return s.map(lambda x: mapping.get(x, x))


def append_df_to_csv(df, out_path):
    """
    Append a dataframe to an existing CSV file without header.
    """
    df.to_csv(out_path, mode="a", index=False, header=False, encoding="utf-8")


def as_str_set(series):
    """
    Convert a Series to a set of strings.
    """
    return set(series.dropna().astype(str).tolist())
```

```python
# ---------------------------------------------------------
# 4) Rebuild LISTEDSECURITY with correct synthetic amount IDs
# ---------------------------------------------------------
# Step 10A intentionally left HASLASTTRADEDVALUE unchanged.
# This block rewrites the full LISTEDSECURITY table so that
# synthetic listed securities point to synthetic monetary amounts.
# ---------------------------------------------------------

LISTEDSECURITY_out = TARGET_TABLES_DIR / "LISTEDSECURITY.csv"

# Start from the original SF1 rows
LISTEDSECURITY_final = [LISTEDSECURITY_SF1.copy()]

for rep in range(1, replica_count + 1):
    listed_clone = LISTEDSECURITY_SF1.copy()

    # Preserve the same synthetic listed security ID logic used in Step 10A
    listed_id_map = dict(
        zip(
            LISTEDSECURITY_SF1["LISTEDSECURITYID"].astype(str),
            make_synth_id(LISTEDSECURITY_SF1["LISTEDSECURITYID"], TARGET_SCALE, "SEC", rep)
        )
    )

    amount_id_map = dict(
        zip(
            MONETARYAMOUNT["MONETARYAMOUNTID"].astype(str),
            make_synth_id(MONETARYAMOUNT["MONETARYAMOUNTID"], TARGET_SCALE, "AMT", rep)
        )
    )

    listed_clone["LISTEDSECURITYID"] = map_with_fallback(listed_clone["LISTEDSECURITYID"], listed_id_map)
    listed_clone["HASLASTTRADEDVALUE"] = map_with_fallback(listed_clone["HASLASTTRADEDVALUE"], amount_id_map)

    if "HASTICKERSYMBOL" in listed_clone.columns:
        listed_clone["HASTICKERSYMBOL"] = mutate_text_value(listed_clone["HASTICKERSYMBOL"], TARGET_SCALE, rep)

    if "HASLEGALNAME" in listed_clone.columns:
        listed_clone["HASLEGALNAME"] = mutate_text_value(listed_clone["HASLEGALNAME"], TARGET_SCALE, rep)

    LISTEDSECURITY_final.append(listed_clone)

LISTEDSECURITY_final_df = pd.concat(LISTEDSECURITY_final, ignore_index=True)
LISTEDSECURITY_final_df.to_csv(LISTEDSECURITY_out, index=False, encoding="utf-8")

print("Rebuilt LISTEDSECURITY successfully.")
print("- rows:", len(LISTEDSECURITY_final_df))
```

```python
# ---------------------------------------------------------
# 6) Generate and append synthetic replicas for the account-rooted side
# ---------------------------------------------------------
# This block handles:
# - PERSON
# - FINANCIALSERVICEACCOUNT
# - HOLDING
# - SECURITIESTRANSACTION
# - MONETARYAMOUNT
# ---------------------------------------------------------

replica_stats = []

for rep in range(1, replica_count + 1):
    print(f"\nGenerating account-rooted replica {rep}/{replica_count} for {TARGET_SCALE}...")

    person_id_map = dict(
        zip(
            PERSON["PERSONID"].astype(str),
            make_synth_id(PERSON["PERSONID"], TARGET_SCALE, "PERS", rep)
        )
    )

    account_id_map = dict(
        zip(
            FINANCIALSERVICEACCOUNT["FINANCIALSERVICEACCOUNTID"].astype(str),
            make_synth_id(FINANCIALSERVICEACCOUNT["FINANCIALSERVICEACCOUNTID"], TARGET_SCALE, "ACCT", rep)
        )
    )

    listed_id_map = dict(
        zip(
            LISTEDSECURITY_SF1["LISTEDSECURITYID"].astype(str),
            make_synth_id(LISTEDSECURITY_SF1["LISTEDSECURITYID"], TARGET_SCALE, "SEC", rep)
        )
    )

    amount_id_map = dict(
        zip(
            MONETARYAMOUNT["MONETARYAMOUNTID"].astype(str),
            make_synth_id(MONETARYAMOUNT["MONETARYAMOUNTID"], TARGET_SCALE, "AMT", rep)
        )
    )

    # -------------------------
    # PERSON
    # -------------------------
    person_clone = PERSON.copy()
    person_clone["PERSONID"] = map_with_fallback(person_clone["PERSONID"], person_id_map)

    if "HASPERSONNAME" in person_clone.columns:
        person_clone["HASPERSONNAME"] = mutate_text_value(person_clone["HASPERSONNAME"], TARGET_SCALE, rep)
    if "HASLASTNAME" in person_clone.columns:
        person_clone["HASLASTNAME"] = mutate_text_value(person_clone["HASLASTNAME"], TARGET_SCALE, rep)
    if "HASFIRSTNAME" in person_clone.columns:
        person_clone["HASFIRSTNAME"] = mutate_text_value(person_clone["HASFIRSTNAME"], TARGET_SCALE, rep)
    if "HASMIDDLENAMEORINITIAL" in person_clone.columns:
        person_clone["HASMIDDLENAMEORINITIAL"] = mutate_text_value(person_clone["HASMIDDLENAMEORINITIAL"], TARGET_SCALE, rep)

    append_df_to_csv(person_clone, PERSON_out)

    # -------------------------
    # FINANCIALSERVICEACCOUNT
    # -------------------------
    account_clone = FINANCIALSERVICEACCOUNT.copy()
    account_clone["FINANCIALSERVICEACCOUNTID"] = map_with_fallback(
        account_clone["FINANCIALSERVICEACCOUNTID"], account_id_map
    )
    account_clone["ISOWNEDBY"] = map_with_fallback(account_clone["ISOWNEDBY"], person_id_map)
    account_clone["ISMANAGEDBY"] = map_with_fallback(account_clone["ISMANAGEDBY"], person_id_map)

    append_df_to_csv(account_clone, FINANCIALSERVICEACCOUNT_out)

    # -------------------------
    # HOLDING
    # -------------------------
    holding_clone = HOLDING.copy()
    holding_clone["HOLDINGID"] = make_synth_id(holding_clone["HOLDINGID"], TARGET_SCALE, "HOLD", rep)
    holding_clone["ISHELDBY"] = map_with_fallback(holding_clone["ISHELDBY"], account_id_map)
    holding_clone["REFERSTO"] = map_with_fallback(holding_clone["REFERSTO"], listed_id_map)

    if "HASDESCRIPTION" in holding_clone.columns:
        holding_clone["HASDESCRIPTION"] = mutate_text_value(holding_clone["HASDESCRIPTION"], TARGET_SCALE, rep)

    append_df_to_csv(holding_clone, HOLDING_out)

    # -------------------------
    # SECURITIESTRANSACTION
    # -------------------------
    txn_clone = SECURITIESTRANSACTION.copy()
    txn_clone["SECURITIESTRANSACTIONID"] = make_synth_id(
        txn_clone["SECURITIESTRANSACTIONID"], TARGET_SCALE, "TXN", rep
    )
    txn_clone["ISFACILITATEDBY"] = map_with_fallback(txn_clone["ISFACILITATEDBY"], account_id_map)
    txn_clone["HASPRICE"] = map_with_fallback(txn_clone["HASPRICE"], amount_id_map)
    txn_clone["REFERSTO"] = map_with_fallback(txn_clone["REFERSTO"], listed_id_map)

    append_df_to_csv(txn_clone, SECURITIESTRANSACTION_out)

    # -------------------------
    # MONETARYAMOUNT
    # -------------------------
    amount_clone = MONETARYAMOUNT.copy()
    amount_clone["MONETARYAMOUNTID"] = map_with_fallback(amount_clone["MONETARYAMOUNTID"], amount_id_map)

    append_df_to_csv(amount_clone, MONETARYAMOUNT_out)

    replica_stats.extend([
        [TARGET_SCALE, rep, "PERSON", len(person_clone)],
        [TARGET_SCALE, rep, "FINANCIALSERVICEACCOUNT", len(account_clone)],
        [TARGET_SCALE, rep, "HOLDING", len(holding_clone)],
        [TARGET_SCALE, rep, "SECURITIESTRANSACTION", len(txn_clone)],
        [TARGET_SCALE, rep, "MONETARYAMOUNT", len(amount_clone)],
    ])

replica_stats_df = pd.DataFrame(replica_stats, columns=[
    "scale_name", "replica_index", "physical_table", "n_rows_appended"
])

replica_stats_df.head(20)
```

```python
# ---------------------------------------------------------
# 8) Validate referential consistency of the full scaled package
# ---------------------------------------------------------

scaled_tables = {}
for table_name in final_tables:
    scaled_tables[table_name] = pd.read_csv(
        TARGET_TABLES_DIR / f"{table_name}.csv",
        dtype=str,
        keep_default_na=False
    )

def missing_fk_count(child_df, child_fk, parent_df, parent_pk):
    child_values = set(child_df[child_fk].dropna().astype(str).unique())
    parent_values = set(parent_df[parent_pk].dropna().astype(str).unique())
    return len(child_values - parent_values)

fk_checks = [
    ("CORPORATION", "ISCLASSIFIEDBY", "INDUSTRYSECTORCLASSIFIER", "INDUSTRYSECTORCLASSIFIERID"),
    ("CORPORATION", "ISDOMICILEDIN", "COUNTRY", "COUNTRYID"),
    ("FINANCIALREPORT", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("ELEMENTSOFFINANCIALREPORT", "ISMEMBEROF", "FINANCIALREPORT", "FINANCIALREPORTID"),
    ("ELEMENTOFFINANCIALSTATEMENT", "ELEMENTOFFINANCIALSTATEMENTID", "ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID"),
    ("DISCLOSURE", "DISCLOSUREID", "ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID"),
    ("SECURITY", "ISPROVIDEDBY", "CORPORATION", "CORPORATIONID"),
    ("LISTEDSECURITY", "LISTEDSECURITYID", "SECURITY", "SECURITYID"),
    ("LISTEDSECURITY", "HASLASTTRADEDVALUE", "MONETARYAMOUNT", "MONETARYAMOUNTID"),
    ("FINANCIALSERVICEACCOUNT", "ISOWNEDBY", "PERSON", "PERSONID"),
    ("FINANCIALSERVICEACCOUNT", "ISMANAGEDBY", "PERSON", "PERSONID"),
    ("HOLDING", "ISHELDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("HOLDING", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "ISFACILITATEDBY", "FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID"),
    ("SECURITIESTRANSACTION", "REFERSTO", "LISTEDSECURITY", "LISTEDSECURITYID"),
    ("SECURITIESTRANSACTION", "HASPRICE", "MONETARYAMOUNT", "MONETARYAMOUNTID"),
]

validation_rows = []

for child_table, child_fk, parent_table, parent_pk in fk_checks:
    missing_count = missing_fk_count(
        scaled_tables[child_table], child_fk,
        scaled_tables[parent_table], parent_pk
    )

    validation_rows.append({
        "child_table": child_table,
        "child_fk": child_fk,
        "parent_table": parent_table,
        "parent_pk": parent_pk,
        "missing_distinct_values": missing_count,
        "fk_ok": (missing_count == 0)
    })

final_scaled_fk_validation = pd.DataFrame(validation_rows)

final_scaled_fk_validation_path = TARGET_VALIDATION_DIR / "final_scaled_fk_validation.csv"
final_scaled_fk_validation.to_csv(final_scaled_fk_validation_path, index=False)

print("Saved:", final_scaled_fk_validation_path.resolve())
final_scaled_fk_validation
```

```python
# ---------------------------------------------------------
# 3) Helper function to profile relationship cardinalities
#    directly from CSV files
# ---------------------------------------------------------
# This avoids loading the full scaled datasets into memory.
# Only the relevant columns are loaded.
# ---------------------------------------------------------

def load_table_columns(table_dir: Path, table_name: str, cols: list[str]) -> pd.DataFrame:
    """
    Load only the required columns of one table.
    """
    file_path = table_dir / f"{table_name}.csv"
    if not file_path.exists():
        raise FileNotFoundError(f"Missing table file: {file_path}")

    return pd.read_csv(file_path, usecols=cols, dtype=str, keep_default_na=False)


def relationship_profile_from_csv(table_dir: Path, parent_table: str, parent_pk: str,
                                  child_table: str, child_fk: str,
                                  relationship_name: str, scale_name: str) -> dict:
    """
    Profile one parent -> child relationship from CSV files.
    """
    parent_df = load_table_columns(table_dir, parent_table, [parent_pk])
    child_df = load_table_columns(table_dir, child_table, [child_fk])

    parent_keys = parent_df[parent_pk].dropna().astype(str)
    child_fk_values = child_df[child_fk].dropna().astype(str)

    counts = child_fk_values.value_counts()
    aligned_counts = parent_keys.map(counts).fillna(0)

    return {
        "scale_name": scale_name,
        "relationship_name": relationship_name,
        "parent_table": parent_table,
        "parent_pk": parent_pk,
        "child_table": child_table,
        "child_fk": child_fk,
        "n_parent_rows": int(len(parent_df)),
        "n_child_rows": int(len(child_df)),
        "min_children_per_parent": float(aligned_counts.min()) if len(aligned_counts) > 0 else None,
        "mean_children_per_parent": round(float(aligned_counts.mean()), 6) if len(aligned_counts) > 0 else None,
        "median_children_per_parent": round(float(aligned_counts.median()), 6) if len(aligned_counts) > 0 else None,
        "p95_children_per_parent": round(float(aligned_counts.quantile(0.95)), 6) if len(aligned_counts) > 0 else None,
        "max_children_per_parent": float(aligned_counts.max()) if len(aligned_counts) > 0 else None,
    }
```

```python
# ---------------------------------------------------------
# 4) Define the table directories for SF1, SF10, and SF30
# ---------------------------------------------------------

scale_dirs = {
    "SF1": BASE_DIR / "sf1_materialized" / "tables",
    "SF10": BASE_DIR / "scaled_corp_rooted" / "SF10" / "tables",
    "SF30": BASE_DIR / "scaled_corp_rooted" / "SF30" / "tables",
}

for scale_name, p in scale_dirs.items():
    if not p.exists():
        raise FileNotFoundError(f"Missing table directory for {scale_name}: {p}")

print("Scale directories confirmed.")
```

```python
# ---------------------------------------------------------
# 5) Profile the main workload-relevant relationships across
#    SF1, SF10, and SF30
# ---------------------------------------------------------

relationship_specs = [
    ("CORPORATION", "CORPORATIONID", "FINANCIALREPORT", "ISPROVIDEDBY", "CORPORATION -> FINANCIALREPORT"),
    ("FINANCIALREPORT", "FINANCIALREPORTID", "ELEMENTSOFFINANCIALREPORT", "ISMEMBEROF", "FINANCIALREPORT -> ELEMENTSOFFINANCIALREPORT"),
    ("ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID", "ELEMENTOFFINANCIALSTATEMENT", "ELEMENTOFFINANCIALSTATEMENTID", "ELEMENTSOFFINANCIALREPORT -> ELEMENTOFFINANCIALSTATEMENT"),
    ("ELEMENTSOFFINANCIALREPORT", "ELEMENTSOFFINANCIALREPORTID", "DISCLOSURE", "DISCLOSUREID", "ELEMENTSOFFINANCIALREPORT -> DISCLOSURE"),
    ("CORPORATION", "CORPORATIONID", "SECURITY", "ISPROVIDEDBY", "CORPORATION -> SECURITY"),
    ("SECURITY", "SECURITYID", "LISTEDSECURITY", "LISTEDSECURITYID", "SECURITY -> LISTEDSECURITY"),
    ("LISTEDSECURITY", "LISTEDSECURITYID", "HOLDING", "REFERSTO", "LISTEDSECURITY -> HOLDING"),
    ("LISTEDSECURITY", "LISTEDSECURITYID", "SECURITIESTRANSACTION", "REFERSTO", "LISTEDSECURITY -> SECURITIESTRANSACTION"),
    ("FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID", "HOLDING", "ISHELDBY", "FINANCIALSERVICEACCOUNT -> HOLDING"),
    ("FINANCIALSERVICEACCOUNT", "FINANCIALSERVICEACCOUNTID", "SECURITIESTRANSACTION", "ISFACILITATEDBY", "FINANCIALSERVICEACCOUNT -> SECURITIESTRANSACTION"),
]

profiles = []

for scale_name, table_dir in scale_dirs.items():
    print(f"Profiling relationships for {scale_name}...")
    for parent_table, parent_pk, child_table, child_fk, rel_name in relationship_specs:
        profiles.append(
            relationship_profile_from_csv(
                table_dir=table_dir,
                parent_table=parent_table,
                parent_pk=parent_pk,
                child_table=child_table,
                child_fk=child_fk,
                relationship_name=rel_name,
                scale_name=scale_name
            )
        )

scale_relationship_profile_comparison = pd.DataFrame(profiles)

scale_relationship_profile_comparison_path = AUDIT_DIR / "scale_relationship_profile_comparison.csv"
scale_relationship_profile_comparison.to_csv(scale_relationship_profile_comparison_path, index=False)

print("Saved:", scale_relationship_profile_comparison_path.resolve())
scale_relationship_profile_comparison
```

```python
# ---------------------------------------------------------
# 6) Compute preservation deltas relative to SF1
# ---------------------------------------------------------
# This quantifies how much the relationship cardinality
# statistics changed from SF1 to the larger scale factors.
# ---------------------------------------------------------

metrics_to_compare = [
    "mean_children_per_parent",
    "median_children_per_parent",
    "p95_children_per_parent",
    "max_children_per_parent"
]

sf1_profiles = scale_relationship_profile_comparison[
    scale_relationship_profile_comparison["scale_name"] == "SF1"
].copy()

comparison_rows = []

for scale_name in ["SF10", "SF30"]:
    scaled_profiles = scale_relationship_profile_comparison[
        scale_relationship_profile_comparison["scale_name"] == scale_name
    ].copy()

    merged = sf1_profiles.merge(
        scaled_profiles,
        on="relationship_name",
        suffixes=("_sf1", f"_{scale_name.lower()}")
    )

    for _, row in merged.iterrows():
        out = {
            "scale_name": scale_name,
            "relationship_name": row["relationship_name"]
        }

        for metric in metrics_to_compare:
            sf1_val = row[f"{metric}_sf1"]
            scaled_val = row[f"{metric}_{scale_name.lower()}"]

            out[f"{metric}_sf1"] = sf1_val
            out[f"{metric}_{scale_name.lower()}"] = scaled_val

            if pd.notna(sf1_val) and sf1_val != 0:
                out[f"{metric}_relative_delta"] = round((scaled_val - sf1_val) / sf1_val, 6)
            elif pd.notna(sf1_val) and sf1_val == 0:
                out[f"{metric}_relative_delta"] = 0.0 if scaled_val == 0 else None
            else:
                out[f"{metric}_relative_delta"] = None

        comparison_rows.append(out)

scale_preservation_deltas = pd.DataFrame(comparison_rows)

scale_preservation_deltas_path = AUDIT_DIR / "scale_preservation_deltas.csv"
scale_preservation_deltas.to_csv(scale_preservation_deltas_path, index=False)

print("Saved:", scale_preservation_deltas_path.resolve())
scale_preservation_deltas.head(20)
```

```python
# ---------------------------------------------------------
# 7) Build a compact preservation summary
# ---------------------------------------------------------

delta_cols = [c for c in scale_preservation_deltas.columns if c.endswith("_relative_delta")]

preservation_summary_rows = []
for scale_name in ["SF10", "SF30"]:
    sub = scale_preservation_deltas[scale_preservation_deltas["scale_name"] == scale_name]

    abs_deltas = []
    for col in delta_cols:
        abs_deltas.extend(sub[col].dropna().abs().tolist())

    mean_abs_delta = round(sum(abs_deltas) / len(abs_deltas), 6) if abs_deltas else None
    max_abs_delta = round(max(abs_deltas), 6) if abs_deltas else None

    preservation_summary_rows.append({
        "scale_name": scale_name,
        "mean_absolute_relative_delta": mean_abs_delta,
        "max_absolute_relative_delta": max_abs_delta
    })

preservation_summary = pd.DataFrame(preservation_summary_rows)
preservation_summary
```
